## 6. Improve results.


<a href="https://docs.pinecone.io/guides/search/rerank-results"> Reranking results </a> is one of the most effective ways to improve search accuracy and relevance, but there are many other techniques to consider. For example:

<a href="https://docs.pinecone.io/guides/search/filter-by-metadata"> Filtering by metadata </a>: When records contain additional metadata, you can limit the search to records matching a filter expression.

<a href="https://docs.pinecone.io/guides/search/hybrid-search"> Hybrid search </a>: You can add lexical search to capture precise keyword matches (e.g., product SKUs, email addresses, domain-specific terms) in addition to semantic matches.

<a href="https://www.pinecone.io/learn/chunking-strategies/"> Chunking strategies </a>: You can chunk your content in different ways to get better results. Consider factors like the length of the content, the complexity of queries, and how results will be used in your application.
