
## The sequence to remember

For this kind of agent program, the sequence is:

**1. Prepare**

* load model
* define tools
* create initial messages

**2. Ask model**

* send `messages` to the LLM

**3. Check response**

* did it return a final answer?
* or did it request a tool?

**4. If tool requested**

* extract tool name, args, id
* run tool
* add tool result back as `ToolMessage`

**5. Repeat**

* send updated messages again

**6. Stop**

* when there are no tool calls, return final answer

That is the whole pattern.

## Another way: remember the roles

This helps a lot:

* **LLM** = decides next step
* **Tools** = do actual work
* **messages** = memory
* **loop** = coordinator
* **ToolMessage** = feedback to the LLM

So when you get lost, ask:

* Who is deciding? → LLM
* Who is doing? → tool
* Who is remembering? → messages
* Who is repeating? → loop

## Why sequence feels confusing

Because the code is not a simple straight line. It is a **conversation loop**:

1. user asks
2. model responds
3. maybe tool runs
4. result goes back to model
5. model continues

So instead of reading it like a normal function, read it like a **call-and-response cycle**.

---

## A practical memory template

For any similar program, keep this template in mind:

```python
define tools
initialize model
bind tools to model
create messages

loop:
    call model with messages
    if no tool call:
        return final answer

    extract tool request
    run tool
    append tool result to messages
```

If you remember just this skeleton, most agent code becomes easier.

## A very useful shortcut

When reading the code, ask these 4 questions in order:

1. **What is the input to the model?**

   * `messages`

2. **What does the model return?**

   * `ai_message`

3. **How do we know whether to use a tool?**

   * `ai_message.tool_calls`

4. **How does the tool result go back to the model?**

   * `ToolMessage(...)`

If you can answer those 4, you understand the sequence.
